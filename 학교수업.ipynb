```python
while True:
    try:
      integer=int(input("1 이상의 정수를 입력하세요.: "))
      while integer<1:
            print("잘못 입력하셨습니다.")
            integer=int(input("1 이상의 정수를 입력하세요.: "))
      print("입력된 정수 : ",integer)

      prime_number=[]

      if integer != 1 :
        for i in range(2,integer+1):
            nCount=0
            for j in range(2,i):
                if i % j == 0:
                    nCount += 1
                    break
            if nCount == 0:
                    prime_number.append(i)

      print("자연수", integer, "이하의 소수는", len(prime_number),"개이며 다음과 같습니다. : ")
      print(prime_number)
      break
    except:
       print("잘못 입력하셨습니다.")
       print("정수를 입력하세요.")
```
